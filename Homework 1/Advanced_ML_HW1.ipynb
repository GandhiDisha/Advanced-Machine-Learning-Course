### 1. What is the difference between R-square and adjusted R square and why is it desirable to use the adjusted value? (4 pts)
Answer: R square error tells us how much <b>variance in the predicted result is explained by the model</b>. Whereas adjusted R square error uses the same concept with the only <b>difference being in the number of predicting variables being considered</b>. R square error considers all the response variables and hence gets manipulated by all too. </b>It can never be negative and just increases with the number of predictor variables being considered</b> whereas adjusted R square error considers <b>only those predictor variables which are essential/significant</b> and hence it might decrease or become negative too with the addition of more variables if they are not significant.
When having <b>multiple predictor variables, it is wise to look at adjusted R square</b> than R square error because the former one is an accurate measure considering the fact that it simply does not increase upon addition of a new variable and also considers the significance of that variable.
For example, if you add a new independent variable to your linear regression model that has no effect on predicting the response variable at all, then this may have no impact on adjusted R square although the R square value can increase. <b>So if you only look at R square, it will cause an illusion that the model is getting better whereas it is not.</b>


### 2. Overfitting usually happens in complex models. Linear Regression is a fairly simple model. Could overfitting happen in Linear Regression? If so, please explain the scenario in which it could happen and how we can tackle it. (6 pts)
Answer:
Yes it is very true that overfitting happens in complex models and Linear Regression is a fairly simple model but there are situations where even <b>Linear Regression can become complex and witness overfitting.</b>
The situation where either the <b>number of predicting variables is equal to the number of observations/data points</b> or when the <b>number of predicting variables is way greater than the number of observations/data points.</b> In such situations the model will try hard to find decision boundary on the training data but the patterns it will observe won't be accurate to fit on unseen data i.e the observations are way too less to come to a generic conclusion about the data. The high number of features will further make it more complex to fit a straight line.
We can tackle the above situation via any of the four ways:

1) Have a large dataset that can be fit in the model and give good accuracy measures.

2) Keep the model simple by only having features which are essential. We can even try feature engineering to consider significant set of features.

3) Use cross validation techniques to just assess how overfit your model is. It won't remove the overfitting at all but will just help you to find out what margin you have to make the model perfect.

4) Use regularization technique such as LASSO, RIDGE to find out ideal set of predicting variables and train your model accordingly.

### 1) Split the data into a training set (75% of data) and a test set (25% of data), using the train_test_split function with random_state = 50. Then scale the data (not including target) so that each of the independent variables would have zero mean and unit variance. You can use the sklearn.preprocessing.scale function for this. Print the first 5 rows of the training set after scaling. (2 pts) 

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.25, random_state=50)
print("Original data Size: ",Y.shape[0])
print("Test data size: ",Y_test.shape[0])
print("Training data size: ",Y_train.shape[0])
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_tr= scaler.fit_transform(X_train)

X_test_tr = scaler.transform(X_test)

X_train_tr[0:5]
```

### 2) Use sklearn.linear_model.Lasso and sklearn.linear_model.Ridge classes to do a 5-fold cross validation using sklearn's KFold. For the sweep of the regularization parameter, we will look at a grid of values ranging from  =10^10  to  =10^6 .

In Python, you can consider this range of values as follows: alpha = 10 numpy.linspace(6,-6,100) so that you can generate 100 uniform values between -6 to 6 as power series.

Fit the 2 regression models with scaled data and report the best chosen  based on cross validation as well as the corresponding scoring metric. The cross validation should happen on your training data using MSE as the scoring metric. 

```python
from sklearn.model_selection import KFold

alpha_List = 10**np.linspace(10,-6,100)
cv_list = []
#Lasso Regression
for alpha in alpha_List:
    cv_error = []
    for train_index, test_index in KFold(n_splits=5).split(X_train_tr):
        lasso= linear_model.Lasso(alpha= alpha)
        lasso.fit(X_train_tr[train_index],Y_train.iloc[train_index])
        lasso_pred = lasso.predict(X_train_tr[test_index])
        cv_error.append(mean_squared_error(lasso_pred, Y_train.iloc[test_index]))
    cv_list.append(np.mean(cv_error))
    
cv_pd= pd.DataFrame(cv_list, index=alpha_List, columns=["MSE"])
lasso_alpha = cv_pd["MSE"].idxmin()
lasso_error = cv_pd.loc[lasso_alpha]["MSE"]
print("Lasso Regression with Cross-Validation:\n\toptimal \u03B1 = {alpha}\n\tMSE: {mse}".format(alpha = lasso_alpha, mse = lasso_error))
```

```python
# Ridge Regression

cv_list = []

for alpha in alpha_List:
    cv_error = []
    for train_index, test_index in KFold(n_splits=5).split(X_train_tr):
        ridge = linear_model.Ridge(alpha= alpha)
        ridge.fit(X_train_tr[train_index],Y_train.iloc[train_index])
        ridge_pred = ridge.predict(X_train_tr[test_index])
        cv_error.append(mean_squared_error(ridge_pred, Y_train.iloc[test_index]))
    
    cv_list.append(np.mean(cv_error))
    
cv_pd = pd.DataFrame(cv_list, index=alpha_List, columns=["MSE"])
ridge_alpha = cv_pd["MSE"].idxmin()
ridge_error = cv_pd.loc[ridge_alpha]["MSE"]
print("Ridge Regression with Cross-Validation:\n\toptimal \u03B1 = {alpha}\n\tMSE: {mse}".format(alpha = ridge_alpha, mse = ridge_error))
```

# Lasso Regression

Acronym for Least Absolute Shrinkage Selector Operator
Lasso assigns a penalty (Loss Function) equal to the MSE + the absolute sum of all coefficients

## Pros:
1.Assigns ZERO weight to coefficients. Hence,model complexity is reduced and overfitting is prevented.
2.Converges faster.
## Cons:
1.Bias in the model is increased.

2.Multicolineraity (or, correlated dependent variables) can not be dealt using LASSO.

# Ridge Regression
Similar to Lasso. However, assigns squared magnitude of the coefficients to the loss function
 
 
## Pros:
1.Ridge is more versatile than LASSO.

2.Ridge does not discard variables randomly by setting them to Zero.

## Cons:
1.Ridge models are more difficult to interpret.

2.Feature coefficients are set to Zero. Hence, ridge retains  insignificant features. This also leads to  increase in complexity and variance of the model.

# ElasticNet

Middle ground between Lasso and ridge regression.
Loss function includes both the absolute sum and squared sum of coefficients.
 
## Pros
1.ElasticNet converges faster than ridge.

2.Eliminates irrelavant features and thereby prevents overfitting.

3.Impact of features which are not affecting output varibale is reduced. (They are not set to 0).
## Cons
1.Loss fucntion is more complex.

2.More compuational power is needed as it is more computationally intesive than lasso and ridge.


### 7) Why did we have to scale the data in ridge and lasso regression? (3 pts)

The primary goal of scaling variables before regularisation is to make sure that the  term regularises each feature equally.

For instance, Let's assume that we have the weights(in grams) and the age of individuals. The age will be in the order of 10^1  while the weight will be in the order of 10^4 as its in gram . If we fit OLS with regularisation on this dataset,it will result in skewing the model coefficients towards either of the features. The output of the OLS model fitted on this dataset would be irrelevant because it would not consider the true relationship between two given features (This is beacuse of the huge disparity in the range and order).

However,scaling of data will result in normalisation(standard normalization) of the values. Now both weight and age will range from -1 to 1 and will be centered around 0 (given, standard normalization scale is used). This scaled dataset will enable regularisation term  to "push-down" coefficient values more evenly. Hence, it will further allow Regression algorithm to work efficiently and converge to the optimal solution.
